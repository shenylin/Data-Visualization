```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
import ipywidgets
```

```python
bf = pd.read_csv('https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_bcubcg_fall2022/main/data/bfro_reports_fall2022.csv',
                 parse_dates = ["date"])                 
```

## Plot 1

```python
bf.columns
```

```python
from matplotlib.lines import Line2D
markersDict = Line2D.markers
markersDict.keys()
markers = []
for m in markersDict:
    if 'None' not in str(m) and len(str(m).strip())>0:
        markers.append(m)
markers[:10]
```

```python
bf.columns.subset = ['county', 'state', 'season', 
       'latitude', 'longitude', 'number', 'classification', 
       'temperature_high', 'temperature_mid', 'temperature_low', 'dew_point',
       'humidity', 'cloud_cover', 'moon_phase', 'precip_intensity',
       'precip_probability', 'pressure', 'uv_index',
       'visibility', 'wind_bearing', 'wind_speed']
```

```python
color_var_subset = ['latitude', 'longitude', 'number', 'dew_point', 'uv_index', 
                    'visibility', 'humidity', 'cloud_cover', 'pressure']
```

```python
bf['humidity']
```

```python
@ipywidgets.interact(x = bf.columns.subset, 
                     y = bf.columns.subset, 
                     color_variable = color_var_subset, 
                     marker = markers) 
def bf_plot(x, y, color_variable, marker):
    fig, ax = plt.subplots(1,1, figsize=(10,5))
    plt.scatter(bf[x], bf[y], marker = marker, c=bf[color_variable], cmap='rainbow')
plt.show()
```

### Write-up

This is an interactive scatter plot that reflects user's selection on x and y field. Originally, the x & y dropdowns are designed to include all the columns (bf.columns) for users to pick. Because even there are some qualitative data (text) that might not make total sense making into a scatter plot, it is still more fun to make it this way so users are able to try out themselves. Personally, I would like to have the option to go through the "trial and error" stage before I decide something does not make sense to me. 

However, after getting a lot system errors, I decided to remove all the qualitative columns because they are not suitable to present in this kind of visualization. Now only numerical and categorical columns are included as options (bf.columns.subset). Yay no more errors! 

Since the dropdown has supported the selection of all columns, it supports the selection of at least 1 categorical column and at least 1 numerical column in for each of x and y.  

In addtion, users get to choose the marker type at will from the third dropdown "marker". I continue the choice of using the variable "pressure" to determine the marker color. Also, this is designed to use a fixed colormap and therefore users can not change the colormap accordingly.  

## Plot 2

```python
@ipywidgets.interact(binned_by = bf.columns.subset, number_of_bins = np.arange(1,9))

def bf_scatter_plot(binned_by, number_of_bins):
    fig, ax = plt.subplots(1,1, figsize=(10,7))
    
    # if NOT numerical data
    if bf[binned_by].dtype != 'float64' and bf[binned_by].dtype != 'int64':
            myseries = bf.groupby(binned_by)[binned_by].count()
            x = myseries.index
            y = myseries.values
            ax.bar(x, y)
    # if numerical data
    else: 
        bf.hist(column=binned_by, ax=ax, bins=number_of_bins)
    
    fig.autofmt_xdate(rotation=90)
    plt.show()
```

### Write-up

I had a hard time understanding what to do with the "aggregate values". At first, I thought we were supposed to stack two variables into one bin. Fortunately, I went to TA's office hour and get the clarification on what we were asked to plot. I then came to a realization on what "aggregation" means when it comes to plotting. It usually means to "groupby" a variable. Therefore, a widget displays aggregate values of a comlumn has been presented. If it is a numerical column, users get to choose the number of presented bins. However, if it is a categorical column, it does not matter what number of presented bin is chosen, it displays the number of existed categories.   

Same design for the second plot, originally the dropdown was designed to include all columns for users to pick because I would like to have users figure out if the histogram makes sense to them. But after getting a lot system errors, I decided to remove all the qualitative columns. Now only numerical and categorical columns are included as options (bf.columns.subset).


Since the binned_by dropdown has supported the selection of all columns, it supports the selection of at least 1 categorical column and at least 1 numerical column.
