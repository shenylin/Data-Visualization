```
#How many people survived?
```

```
len(df[df['Survived'] == 1])
```

```
sur_df = df.groupby('Survived').count()

plt.barh(sur_df['PassengerId'].index,sur_df['PassengerId'].values)
plt.ylabel('Survival of passengers')
plt.yticks(sur_df['PassengerId'].index, ['Died','Lived'])
plt.xlabel('Number of people (hundreds)')
```

```
#Members of what sex survived more?
```

```
df.groupby('Sex')['Survived'].mean()
```

```
sex_df = df.groupby(['Sex','Survived'], as_index=True).count()

width = 0.4
plt.barh(sex_df['PassengerId']['female'].index, sex_df['PassengerId']['female'].values, width, label='Women', color='pink')
plt.barh(sex_df['PassengerId']['male'].index + width, sex_df['PassengerId']['male'].values, width, label='Men', color='blue')
plt.ylabel('Survival of passengers by sex')
plt.yticks(sex_df['PassengerId']['female'].index, ['Died', 'Lived'])
plt.xlabel('Number of people (hundreds)')
plt.legend()
```

```
sex_df.index
```

```
#Did class membership affect survival rates?
```

```
df.groupby('Pclass').mean()
```

```
class_df = df.groupby('Pclass').mean()

plt.barh(class_df.index, class_df['Survived'].values)
plt.ylabel('Survival of passengers by class')
plt.yticks(class_df.index, ['First','Second','Third'])
plt.xlabel('mean')
plt.xticks(np.arange(0.0,0.7,0.1), ['0%','10%','20%','30%','40%','50%','60%'])
```
