```python
# 1. "hardcode" a value that we want to change interactively
# 2. make a function with that hard-coded value as an input
# 2.5 test our function with a few different values
# 3. decorate our function with a ipywidgets.interact
```

```python
# 1. hardcode a value
colormap_name = 'seismic'

plt.imshow(michigan,cmap=colormap_name)
plt.colorbar()
plt.clim(-352,352)
plt.show()
```

```python
# 2. make a function with that hard-coded value as an input
def make_michigan_plot(colormap_name):
    plt.imshow(michigan,cmap=colormap_name)
    plt.colorbar()
    plt.clim(-352,352)
    plt.show()
```

```python
#2.5 test!
make_michigan_plot('seismic')
```

```python
make_michigan_plot('terrain')
```

```python
# 3. decorate our function with a ipywidgets.interact
@ipywidgets.interact(colormap_name=plt.colormaps(), button=True)
def make_michigan_plot(colormap_name, button):
    plt.imshow(michigan,cmap=colormap_name)
    plt.colorbar()
    plt.clim(-352,352)
    plt.show()
```
