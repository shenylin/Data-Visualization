### Summary

The dataset is full of interesting characteristics to play with! It contains 1688 listings in Newark city, NJ. It also contains 17 characteristics. To summarize, several characteristics that we're going to exploit heavily will be:

- host_name: the name of host of the listing, nominal
- neighbourhood: the neighborhood of the listing, nominal
- latitude/longitude: we can plot a map using these data, geodata
- room_type: the type of the listing, whether it is an entire home/apt, a privite room or a shared room, nominal
- price: the price of the listing, numerical
- minimum_nights: the minimum nights stayed in the property, numerical
- number_of_reviews: the number of reviews each listings get over the years listed, numerical
- last_review: date of the last review received as of 28 September, 2022, temporal/ordinal
- availability_365: the available days in a year, numerical

look at statistics of our subset from different years using "describe" as we've done before for the full dataset.  In this way we can "mutate" our data:

Plot the **maximum** price of any listing created in a particular type of room:

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.plot(stats["max"], marker='.', linewidth = 1.0)
ax.set_xlabel("room_type")
ax.set_ylabel("price")
plt.show()
```

We can make a similar plot for the minimum value of the price of listing in each room type:

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.plot(stats["min"], marker='.', linewidth = 1.0)
ax.set_xlabel("room_type")
ax.set_ylabel("price")
plt.show()
```

# Plotting styles and default plotting colormaps

Let's now plot both min and max on the same plot:

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.plot(stats["max"], marker='.', linewidth = 1.0)
ax.plot(stats["min"], marker='.', linewidth = 1.0)
ax.set_xlabel("room_type")
ax.set_ylabel("price")
ax.set_yscale("log")
plt.show()
```

Add a legend:

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.plot(stats["max"], marker='.', linewidth = 1.0, label="Max")
ax.plot(stats["min"], marker='.', linewidth = 1.0, label="Min")
ax.set_xlabel("room_type")
ax.set_ylabel("price")
ax.set_yscale("log")

# Note the difference between who "owns" this legend
#fig.legend()
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.plot(stats["max"], marker='.', linewidth = 1.0, label="Max")
ax.plot(stats["min"], marker='.', linewidth = 1.0, label="Min")
ax.set_xlabel("room_type")
ax.set_ylabel("price")
ax.set_yscale("log")
ax.fill_between(stats.index, stats["min"], stats["max"], color = "#dddddd")

ax.legend()

plt.show()
```

### Plotting Style

"ggplot" package with Python

```python
with plt.style.context("ggplot"):
    fig, ax = plt.subplots(figsize=(8,6))
    ax.plot(stats["max"], marker='.', linewidth = 1.0, label = "Max")
    ax.plot(stats["min"], marker='.', linewidth = 1.0, label = "Min")
    ax.fill_between(stats.index, stats["min"], stats["max"], color = "lightgray")
    ax.set_xlabel("room_type")
    ax.set_ylabel("price")
    ax.set_yscale("log")
    ax.legend()
plt.show()
```

### The `with` statement

```python
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(stats["max"], marker='.', linewidth = 1.0, label = "Max")
ax.plot(stats["min"], marker='.', linewidth = 1.0, label = "Min")
ax.fill_between(stats.index, stats["min"], stats["max"], color = "lightgray")
ax.set_xlabel("room_type")
ax.set_ylabel("price")
ax.set_yscale("log")
ax.legend()

plt.show()
```

But what if we want to see how our plot would look with each of these styles?  We could just make a bunch of plots OR we can play with this interactively with the `ipywidgets` library.

We'll talk about `ipywidgets` extensively starting next week, but we can dip our toes into this package now and start getting a feel for how they work.

```python
import ipywidgets
```

First, lets tell jupyter `ipywidgets` that we want to mess around with the style of the plot:

```python
@ipywidgets.interact(style = plt.style.available)
def make_plot(style):
    with plt.style.context(style):
        fig, ax = plt.subplots(figsize=(8,6))
        ax.plot(stats["max"], marker='.', linewidth = 1.0, label = "Max")
        ax.plot(stats["min"], marker='.', linewidth = 1.0, label = "Min")
        ax.fill_between(stats.index, stats["min"], stats["max"], color = "lightgray")
        ax.set_xlabel("room_type")
        ax.set_ylabel("price")
        ax.set_yscale("log")
        ax.legend()        
# so now you can see that we get a little dropdown menu that lists
```
