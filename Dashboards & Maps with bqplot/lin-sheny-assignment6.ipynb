```python
def generate_histogram_from_issueyear(lic, nissueyear=10, nmedianeffeyear=10, issueyearmin=1900, issueyearmax=2020, medianeffedaysmin=0, medianeffedaysmax=40000, takeLog=True):
    issueyear_bins = np.linspace(issueyearmin, issueyearmax, nissueyear+1)
    medianeffedays_bins = np.linspace(medianeffedaysmin, medianeffedaysmax, nmedianeffedays+1)
    hist2d, issueyear_edges, medianeffedays_edges = np.histogram2d(lic['Original Issue Year'], 
                                                   lic['Median_Effective_Days'], 
                                                   weights=lic['Effective Days'],
                                                  bins = [issueyear_bins,medianeffedays_bins])
    hist2d = hist2d.T
    if takeLog:
        hist2d[hist2d <= 0] = np.nan # set zeros to NaNs
        # then take log
        hist2d = np.log10(hist2d)
    issueyear_centers = (issueyear_edges[:-1] + issueyear_edges[1:]) / 2
    medianeffedays_centers = (medianeffedays_edges[:-1] + medianeffedays_edges[1:]) / 2
    return hist2d, issueyear_centers, medianeffedays_centers, issueyear_edges, medianeffedays_edges
```

```python
# data
hist2d, issueyear_centers, medianeffedays_centers, issueyear_edges, medianeffedays_edges = generate_histogram_from_issueyear(lic)
```
